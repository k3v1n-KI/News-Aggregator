```python
from datetime import datetime

def toDateTime(date_string):
    date_processing = date_string.replace(
        'T', '-').replace(':', '-').replace("Z", "").split('-')
    date_processing = [int(v) for v in date_processing]
    date_out = datetime(*date_processing)
    return date_out
```
